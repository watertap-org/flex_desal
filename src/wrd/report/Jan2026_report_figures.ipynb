```python
# PRO 1 Power Use and Permeate Production
# Data Points: 8/19/2024 and 3/13/2024
measured_power_8_19 = 196.25
modeled_power_8_19 =  170.1
DT_power_8_19 = 203.0

measured_perm_8_19 = 1608.2
modeled_perm_8_19 =  1436.8
DT_perm_8_19 = 1602.4

measured_power_3_13 = 189.6
modeled_power_3_13 =  162.1
DT_power_3_13 = 203.9

measured_perm_3_13 = 1404.7
modeled_perm_3_13 =  1432
DT_perm_3_13 = 1468

fig_2, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 12))

# Subplot 1: 8/19 Power data
categories_1 = ['Measured', 'Modeled', 'Digital Twin']
values_1 = [measured_power_8_19, modeled_power_8_19, DT_power_8_19]
ax1.bar(categories_1, values_1, color=['coral', 'steelblue', 'green'])
eb1 = ax1.errorbar(0,measured_power_8_19,yerr=measured_power_8_19 * 0.15,  color='black', capsize=15, linewidth=2)
eb1[-1][0].set_linestyle('--')
ax1.set_ylabel('Power (kW)', fontsize=14)
ax1.set_title('PRO 1 Power - 8/19', fontsize=14)
ax1.set_ylim([0, max(values_1) * 1.25])
ax1.tick_params(labelsize=14)

# Subplot 2: 3/13 Power data
categories_2 = ['Measured', 'Modeled', 'Digital Twin']
values_2 = [measured_power_3_13, modeled_power_3_13, DT_power_3_13]
ax2.bar(categories_2, values_2, color=['coral', 'steelblue', 'green'])
eb2 = ax2.errorbar(0,measured_power_3_13,yerr=measured_power_3_13 * 0.15,  color='black', capsize=15, linewidth=2)
eb2[-1][0].set_linestyle('--')
ax2.set_ylabel('Power (kW)', fontsize=14)
ax2.set_title('PRO 1 Power - 3/13', fontsize=14)
ax2.set_ylim([0, max(values_2) * 1.25])
ax2.tick_params(labelsize=14)

# Subplot 3: 8/19 Permeate data
categories_3 = ['Measured', 'Modeled', 'Digital Twin']
values_3 = [measured_perm_8_19, modeled_perm_8_19, DT_perm_8_19]
ax3.bar(categories_3, values_3, color=['coral', 'steelblue', 'green'])
eb3 = ax3.errorbar(0, measured_perm_8_19, yerr=measured_perm_8_19 * 0.15, color='black', capsize=15, linewidth=2)
eb3[-1][0].set_linestyle('--')
ax3.set_ylabel('Permeate Flow (gpm)', fontsize=14)
ax3.set_title('PRO 1 Permeate - 8/19', fontsize=14)
ax3.set_ylim([0, max(values_3) * 1.25])
ax3.tick_params(labelsize=14)

# Subplot 4: 3/13 Permeate data
categories_4 = ['Measured', 'Modeled', 'Digital Twin']
values_4 = [measured_perm_3_13, modeled_perm_3_13, DT_perm_3_13]
ax4.bar(categories_4, values_4, color=['coral', 'steelblue', 'green'])
eb4 = ax4.errorbar(0, measured_perm_3_13, yerr=measured_perm_3_13 * 0.15, color='black', capsize=15, linewidth=2)
eb4[-1][0].set_linestyle('--')
ax4.set_ylabel('Permeate Flow (gpm)', fontsize=14)
ax4.set_title('PRO 1 Permeate - 3/13', fontsize=14)
ax4.set_ylim([0, max(values_4) * 1.25])
ax4.tick_params(labelsize=14)

plt.tight_layout()
plt.savefig('PRO1_Power_Comparison.png', dpi=300)
```
