```python
# UV- AOP Data
# User input to define x and y data
plot_title = "UV AOP Unit 1 vs. Feed Flowrate"
x_data_name = "UV1_MGD"
y_data_name_list = ["UV1_kW_masked"]

# Derived quantities
Data["UV1_kW_masked"] = Data["UV_train1_kW"]
Data.loc[Data["UV1_MGD"] < 1, "UV1_kW_masked"] = np.nan
Data.loc[Data["UV_train1_kW"] < 0, "UV1_kW_masked"] = np.nan
# Create Plot
x_data = Data[x_data_name]
figure,ax = plt.subplots()
for y_data_name in y_data_name_list:
    y_data = Data[y_data_name]
    plt.plot(x_data,y_data,'o',label="UV1 Power",color='steelblue')
plt.title(plot_title)
plt.xlabel("UV AOP Flowrate (MGD)")
plt.ylabel("Power Consumption (kW)")
# Adjust Axes as needed
plt.xlim(0,8)
# plt.ylim(0,1)

# Add the Trendline
x_fit = np.linspace(0,8)
z = 15.9 * x_fit
ax.plot(x_fit,z,"--k",label="y=15.9*flow",linewidth=2)
plt.legend(loc='upper left')
#Export Figure
plt.savefig("UV_AOP_Unit1_Power_Feb_21.png", dpi=300)
```
